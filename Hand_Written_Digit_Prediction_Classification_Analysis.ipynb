**Hand Written Digit Prediction - Classification Analysis**

```python
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
import numpy as np

# Load dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize the images.
train_images = train_images / 255.0
test_images = test_images / 255.0

# Flatten the images.
train_images_flattened = train_images.reshape(train_images.shape[0], -1)
test_images_flattened = test_images.reshape(test_images.shape[0], -1)
```

```python
import matplotlib.pyplot as plt

# Plot the first few images
for i in range(9):
    plt.subplot(330 + 1 + i)
    plt.imshow(train_images[i], cmap=plt.get_cmap('gray'))
    plt.title(train_labels[i])
plt.show()
```

```python
# This might take a while as Colab installs the required dependencies.
!pip install scikit-learn

from sklearn.linear_model import LogisticRegression

# Create and train a logistic regression model
model = LogisticRegression(max_iter=100, solver='lbfgs', multi_class='multinomial', n_jobs=-1)
model.fit(train_images_flattened, train_labels)
```

```python
# Evaluate the model
score = model.score(test_images_flattened, test_labels)
print("Accuracy:", score)
```

```python
# Make predictions
predictions = model.predict(test_images_flattened[0:5])

# Visualize the predictions along with the actual images
for i in range(5):
    plt.subplot(151 + i)
    plt.imshow(test_images[i], cmap=plt.get_cmap('gray'))
    plt.title(f"Pred: {predictions[i]}")
plt.show()
```
